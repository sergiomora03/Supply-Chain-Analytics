```python
from pulp import *
import numpy as np
```

```python
#Definimos parámetros

max_dem = [50000*(1+0.05)**i for i in range(0,10)]
cost_plant = 16 # Costo por unidad de capacidad para la construcción de la planta
prec_vent = 3 #Precio de venta al publico por unidad
cost_prod = 0.2 #Costo de producción
cost_cap = 0.4 #Costo de capacidad por unidad
```

```python
#Problem definition

prob = LpProblem("wozac",LpMaximize)
Arcs_Def = [i for i in range(0,10)]
```

```python
# Decision Variables (integer variables)

PR = LpVariable.dicts("Production",Arcs_Def,0,None,LpInteger)
CP = LpVariable("Capacity",0,None,LpInteger,)
```

```python
#Objective Function

prob += lpSum([((prec_vent-cost_prod)*PR[a])-(cost_cap*CP)for a in Arcs_Def]) -(CP*(cost_plant)), "Sum of profit" 
```

```python
#Constraints (1)

for a in Arcs_Def:
    prob += PR[a] <= max_dem[a], "Max Dem" + str(a)   

for a in Arcs_Def:
    prob += PR[a] <= CP, "Capacity" + str(a)   
```

```python
#Solve

prob.writeLP("wozac.lp")
prob.solve()
print("Status:", LpStatus[prob.status])
```

```python
#Print optimal objective function value

print("Optimal Profit=", "$"+ str(value(prob.objective)))
```

```python
#Print optimal solution

for variable in prob.variables():
    if variable.varValue > 0:
        print ("{}* = {}".format(variable.name, variable.varValue))
```
